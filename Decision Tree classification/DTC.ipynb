# Decision Tree Classifier

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1 - Read the Train dataset

```python
import pandas as pd 
df = pd.read_csv("iris.csv")
df.head()
```

### Step 2 : Perform Basic Data Quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(species)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

### Step 4: Perform Preprocessing on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Perform Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Step 6: Build and Tune decision tree
1. max_depth
2. min_samples_split
3. min_samples_leaf
4. criterion - gini, entropy

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=1,
                               min_samples_leaf=5,
                               min_samples_split=5,
                               criterion='entropy')
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Hyperparameter tuning the model with gridsearch cv

```python
params = {'max_depth':[2, 3, 4, 5, 6],
          'min_samples_split':[3, 4, 5, 6, 7],
          'min_samples_leaf':[3, 4, 5, 6, 7],
          'criterion':['gini', 'entropy']}
```

```python
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier()
gscv = GridSearchCV(dtc, param_grid=params, cv=5, scoring='f1_macro')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

### Evaluate tuned model

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```

```python
ypred_train = best_dtc.predict(xtrain)
ypred_test = best_dtc.predict(xtest)
```
