# Step 1: Read the train dataset

```python
import pandas as pd 
df = pd.read_csv('Cars93.csv', na_values=['','NA'], keep_default_na=False)
df.head()
```

### Step 2: Perfom basic data quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4: Create a preprocessing pipeline for X

```python
# Seperate cat and con features for X
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('OHE', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 5: Apply train test split
20% - 30% - Unknown to model (Test)

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=60)
```

```python
xtest.shape
```

# Step 6: Model Building and evaluation

```python
from sklearn.linear_model import LinearRegression
model_linear = LinearRegression()
model_linear.fit(xtrain, ytrain)
```

```python
model_linear.score(xtrain, ytrain)
```

```python
model_linear.score(xtest, ytest)
```

# Ridge Model 

```python
from sklearn.linear_model import Ridge 
model_ridge = Ridge(alpha=2)
model_ridge.fit(xtrain, ytrain)
```

```python
model_ridge.score(xtrain, ytrain)
```

```python
model_ridge.score(xtest, ytest)
```

### Tuning the ridge model with GridSearchCV

```python
import numpy as np 
params1 = {'alpha':np.arange(start=0.1, stop=100, step=0.1)}
params1
```

```python
from sklearn.model_selection import GridSearchCV
rr = Ridge()
gscv1 = GridSearchCV(rr, param_grid=params1, cv=5, scoring='neg_mean_squared_error')
gscv1.fit(xtrain, ytrain)
```

```python
gscv1.best_params_
```

```python
gscv1.best_score_
```

```python
best_ridge = gscv1.best_estimator_
best_ridge
```

```python
best_ridge.score(xtrain, ytrain)
```

```python
best_ridge.score(xtest, ytest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(best_ridge, xtrain, ytrain)
```

```python
evaluate_model(best_ridge, xtest, ytest)
```

```python
from sklearn.model_selection import cross_val_score
ridge_scores = cross_val_score(best_ridge, xtrain, ytrain, cv=5, scoring='r2')
ridge_scores
```

```python
ridge_scores.mean()
```
